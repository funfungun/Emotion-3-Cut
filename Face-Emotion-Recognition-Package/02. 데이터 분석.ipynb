### #1. 모듈 선언

```python
from libs.FacialExpression import DataAnalyzer

data_analyzer = DataAnalyzer()
```

### #2. 데이터 플로팅하기

```python
import matplotlib.pyplot as plt

test_data = [1, 2, 3, 4]

plt.plot(test_data)
plt.show()
```

### #3. 데이터 특징 정의하기

```python
def squred(value):
    result = value * value

    return result

data_analyzer.compile(feature_fn=squred)
```

### #4. 데이터 특징 추출후 플로팅 하기

```python
# x: 데이터의 x축 값 (list) - 1차원 데이터일 경우 None을 입력시 자동으로 설정됨
# y: 데이터의 y축 값 (list)
# name: 데이터의 이름 (문자열) - (display 목적) 
# threshold: 임계치로 설정할 값 (숫자) - (display 목적) 
# contrast: 기존 데이터와 비교하여 출력 할 것인지 여부 (True/False)

data_analyzer.feed(x=None, y=test_data, name="Test Data")
```

raw 데이터와 함께 플로팅 하는 것도 가능함

```python
# contrast: True로 설정할 경우 기존 데이터(검은색)도 같이 출력

data_analyzer.feed(x=None, y=test_data, name="Test Data", contrast=True)
```

threshold 값을 설정해주면 적용 결과를 미리 확인할 수 있음

```python
# threshold: 임계치를 설정시 해당 부분을 표시

data_analyzer.feed(x=None, y=test_data, name="Test Data", contrast=True, threshold=8)
```

표정 변화에 따라 두드러진 변화를 보이는 특징을 정의하고, 표정 변화를 잘 인식할 수 있도록 임계치를 설정해보자.

### #5. 두 데이터 비교하기
서로 다른 두 데이터를 플롯하고 싶을 경우, 미리 플롯할 데이터들을 입력주어야 한다.

```python
# 비교 테스트에 이용할 3개의 2차원 데이터를 준비 (0번 인덱스는 x축, 1번 인덱스는 y축 데이터가 되도록 구성할 것)
data1 = [[1, 2, 3, 4], [1, 2, 3, 4]]
data2 = [[1, 2, 3, 4], [2, 4, 6, 8]]
data3 = [[1, 2, 3, 4], [5, 10, 15, 20]]

# append_data(name, value): 비교할 데이터들을 미리 입력한다.
# name: 입력할 데이터의 간단한 이름 (display시 알아보기 위함)
# value: 입력할 데이터 리스트. 2차원 배열이여야 하며 0번 인덱스에 x축 데이터, 1번 인덱스에 y축 데이터가 입력되어야 함.
data_analyzer.appendData(name="data1", value=data1)
data_analyzer.appendData(name="data2", value=data2)
data_analyzer.appendData(name="data3", value=data3)

# 한 번에 플롯하고 싶은 데이터들의 이름을 리스트 형태로 정리하여 입력해준다.
# line
# scatter
data_analyzer.plotDatas(names=['data1', 'data2'], type="line")
```

미리 입력시켜준 데이터들이 어떤 것들이 있는지 간단하게 이름들로 확인할 수 있다.

```python
data_analyzer.printDatas()
```
